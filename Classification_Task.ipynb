## 1. Dataset Description
We will be working with the **Breast Cancer Wisconsin Diagnostic Dataset**.  
- The dataset contains **569 records** of patients.  
- Each record has **30 numerical features** (e.g., radius, texture, perimeter, smoothness, etc.) that describe characteristics of cell nuclei present in breast cancer tissue.  
- The **target variable** is `Diagnosis`, which has two classes:
  - **M (Malignant)**
  - **B (Benign)**  

This is a **binary classification problem**, and we will use **Logistic Regression** to build our model.

```python
# ================================
# ü©∫ Breast Cancer Classification
# Logistic Regression End-to-End
# ================================

# ---- Imports
import sys
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_curve,
    auc,
)

# -------------------------------
# 1) DATASET DESCRIPTION & LOAD
# -------------------------------
print("=== Step 1: Load Dataset ===")

features = [
    "ID number",
    "Diagnosis",
    "radius_mean",
    "texture_mean",
    "perimeter_mean",
    "area_mean",
    "smoothness_mean",
    "compactness_mean",
    "concavity_mean",
    "concave points_mean",
    "symmetry_mean",
    "fractal_dimension_mean",
    "radius_se",
    "texture_se",
    "perimeter_se",
    "area_se",
    "smoothness_se",
    "compactness_se",
    "concavity_se",
    "concave points_se",
    "symmetry_se",
    "fractal_dimension_se",
    "radius_worst",
    "texture_worst",
    "perimeter_worst",
    "area_worst",
    "smoothness_worst",
    "compactness_worst",
    "concavity_worst",
    "concave points_worst",
    "symmetry_worst",
    "fractal_dimension_worst",
]

def load_uci_or_fallback():
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
    try:
        df = pd.read_csv(url, names=features)
        source = "UCI (wdbc.data)"
        # Basic shape checks
        assert df.shape[1] == 32, "Unexpected column count from UCI source."
        return df, source
    except Exception as e:
        print(f"[Info] Could not load UCI file directly ({e}). Falling back to sklearn dataset.")
        from sklearn.datasets import load_breast_cancer
        data = load_breast_cancer(as_frame=True)
        df = data.frame.copy()
        # Align to our schema: create ID, Diagnosis, and 30 features
        df.insert(0, "ID number", np.arange(1, len(df) + 1))
        df.insert(1, "Diagnosis", np.where(data.target == 0, "M", "B"))  # sklearn: 0=malignant, 1=benign
        # Rename columns to approximate our feature naming
        # (keeps everything consistent downstream)
        rename_map = {
            'mean radius':'radius_mean','mean texture':'texture_mean','mean perimeter':'perimeter_mean','mean area':'area_mean',
            'mean smoothness':'smoothness_mean','mean compactness':'compactness_mean','mean concavity':'concavity_mean',
            'mean concave points':'concave points_mean','mean symmetry':'symmetry_mean','mean fractal dimension':'fractal_dimension_mean',
            'radius error':'radius_se','texture error':'texture_se','perimeter error':'perimeter_se','area error':'area_se',
            'smoothness error':'smoothness_se','compactness error':'compactness_se','concavity error':'concavity_se',
            'concave points error':'concave points_se','symmetry error':'symmetry_se','fractal dimension error':'fractal_dimension_se',
            'worst radius':'radius_worst','worst texture':'texture_worst','worst perimeter':'perimeter_worst','worst area':'area_worst',
            'worst smoothness':'smoothness_worst','worst compactness':'compactness_worst','worst concavity':'concavity_worst',
            'worst concave points':'concave points_worst','worst symmetry':'symmetry_worst','worst fractal dimension':'fractal_dimension_worst'
        }
        df = df.rename(columns=rename_map)
        # Ensure column order
        ordered = features + [c for c in df.columns if c not in features]
        df = df[features]
        source = "sklearn.datasets.load_breast_cancer (fallback)"
        return df, source

df, source = load_uci_or_fallback()
print(f"Loaded from: {source}")
print("Dataset shape:", df.shape)
print(df.head(3), "\n")

# 2) DATA INVESTIGATION
print("=== Step 2: Data Investigation ===")
print("\nInfo:")
print(df.info())
print("\nDescribe (numeric):")
print(df.drop(columns=["Diagnosis"], errors="ignore").describe())

print("\nNull values per column:\n", df.isnull().sum())
print("Duplicate rows:", df.duplicated().sum())

# Check class balance
print("\nClass counts:")
print(df["Diagnosis"].value_counts())
print("\nClass proportions:")
print(df["Diagnosis"].value_counts(normalize=True))

# --------------------------------
# 3) DATA VISUALIZATION
# --------------------------------
print("\n=== Step 3: Data Visualization ===")
# 3a) Class distribution bar chart
plt.figure(figsize=(5, 4))
cls_counts = df["Diagnosis"].value_counts()
plt.bar(cls_counts.index, cls_counts.values)
plt.title("Class Distribution (Diagnosis)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 3b) Histograms for a few representative features from mean/se/worst
plot_features = [
    "radius_mean", "texture_mean", "perimeter_mean", "area_mean",
    "radius_se", "texture_se", "perimeter_se", "area_se",
    "radius_worst", "texture_worst", "perimeter_worst", "area_worst",
]
# To avoid a giant wall of plots, we batch them
cols = 3
rows = math.ceil(len(plot_features)/cols)
for i, feat in enumerate(plot_features, 1):
    plt.figure(figsize=(5, 3))
    plt.hist(df[feat], bins=30)
    plt.title(f"Histogram: {feat}")
    plt.xlabel(feat)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

# 3c) Correlation heatmap (subset to keep labels readable)
corr_features = [
    "radius_mean", "texture_mean", "perimeter_mean", "area_mean",
    "smoothness_mean", "compactness_mean", "concavity_mean",
    "concave points_mean", "symmetry_mean"
]
corr = df[corr_features].corr()
plt.figure(figsize=(7, 6))
im = plt.imshow(corr, interpolation='nearest', aspect='auto')
plt.colorbar(im, fraction=0.046, pad=0.04)
plt.xticks(range(len(corr_features)), corr_features, rotation=45, ha="right")
plt.yticks(range(len(corr_features)), corr_features)
plt.title("Correlation Heatmap (subset)")
plt.tight_layout()
plt.show()

# 4) DATA PREPROCESSING

print("\n=== Step 4: Data Preprocessing ===")

df_processed = df.copy()
df_processed["Diagnosis"] = df_processed["Diagnosis"].map({"M": 1, "B": 0})

X = df_processed.drop(columns=["ID number", "Diagnosis"])
y = df_processed["Diagnosis"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Features shape (after drop+scale):", X_scaled.shape)
print("Target shape:", y.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
print("Train size:", X_train.shape[0], " Test size:", X_test.shape[0])

pos_ratio = y.mean()
imbalance = abs(0.5 - pos_ratio)
print(f"Positive class (Malignant=1) proportion overall: {pos_ratio:.3f}")

use_balanced = imbalance > 0.1
print("Use class_weight='balanced':", use_balanced)

# 5) MODEL TRAINING (Logistic Regression)

print("\n=== Step 5: Model Training ===")
model = LogisticRegression(
    max_iter=1000, solver="lbfgs",
    class_weight="balanced" if use_balanced else None
)
model.fit(X_train, y_train)
print("Model trained.")

# 6) MODEL EVALUATION

print("\n=== Step 6: Model Evaluation ===")

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

# 6.1 Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
print("Confusion Matrix:\n", cm)

# Plot CM as heatmap via imshow with annotations
plt.figure(figsize=(4.5, 4))
im = plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix")
plt.colorbar(im, fraction=0.046, pad=0.04)
tick_marks = np.arange(2)
plt.xticks(tick_marks, ["Benign (0)", "Malignant (1)"])
plt.yticks(tick_marks, ["Benign (0)", "Malignant (1)"])
# annotations
thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, str(cm[i, j]),
                 ha="center", va="center",
                 color="white" if cm[i, j] > thresh else "black")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.tight_layout()
plt.show()

# 6.2 Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=3))

# 6.3 ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
print(f"AUC: {roc_auc:.3f}")

plt.figure(figsize=(5, 4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# 7) MODEL INTERPRETATION

print("\n=== Step 7: Model Interpretation ===")
train_acc = model.score(X_train, y_train)
test_acc = model.score(X_test, y_test)
print(f"Training Accuracy: {train_acc:.3f}")
print(f"Testing  Accuracy: {test_acc:.3f}")


####################### Training vs Testing Performance

#1) Training accuracy: close to 100% (usually around 0.98‚Äì1.0 for this dataset).

#2) Testing accuracy: also very high (typically 0.95‚Äì0.98).

#3) The gap between training and testing is small, meaning the model generalizes well.

########################### Overfitting or Underfitting?

#1) Since both training and testing accuracies are high and close together:

#2) No significant overfitting (the model is not memorizing the training set).

#3) No underfitting (the model is capturing the relationships in the data effectively).

#4) Logistic Regression is performing well for this dataset.

###################### Error Analysis

#1) False Positives (FP): The model sometimes predicts Malignant when the tumor is actually Benign.

#2) False Negatives (FN): The model sometimes predicts Benign when the tumor is actually Malignant.

#3) In medical problems, False Negatives are more dangerous than False Positives.

# ####################### Medical Consequences

#1) False Positives (FP):

#2) A patient without cancer is wrongly flagged as malignant.

#3) Consequence: unnecessary stress, further testing, or biopsies.

#1) False Negatives (FN):

#2) A patient with malignant cancer is misclassified as benign.

#3) Consequence: missed early diagnosis, delayed treatment ‚Üí potentially life-threatening.
```
